# Feature Engineering — примеры новых признаков

```python
import pandas as pd
import sys
sys.path.append('..')
from src.features.build_features import build_features

df = pd.read_parquet('../data/processed/train.parquet')
df_feat = build_features(df)

original_cols = df.columns.tolist()
new_cols = [c for c in df_feat.columns if c not in original_cols]
print(f'Новых признаков: {len(new_cols)}')
print(new_cols)
```

## Лаговые признаки

```python
lag_cols = ['user_id', 'date', 'steps', 'steps_lag_1', 'steps_lag_3', 'steps_lag_7']
sample = df_feat[df_feat['user_id'] == df_feat['user_id'].iloc[0]].sort_values('date')
sample[lag_cols].dropna().head(10)
```

## Скользящие средние

```python
roll_cols = ['date', 'steps', 'steps_roll_mean_3', 'steps_roll_mean_7', 'steps_roll_mean_14']
sample[roll_cols].dropna().head(10)
```

## Календарные признаки

```python
cal_cols = ['date', 'day_of_week', 'is_weekend', 'day_of_month', 'month']
df_feat[cal_cols].drop_duplicates('date').sort_values('date').head(10)
```

## Streak активности

```python
streak_cols = ['date', 'steps', 'activity_streak']
sample[streak_cols].dropna().head(15)
```

## Статистика пропусков после feature engineering

```python
missing = df_feat[new_cols].isnull().sum()
missing_pct = (missing / len(df_feat) * 100).round(1)
pd.DataFrame({'missing': missing, 'missing_%': missing_pct})
```
